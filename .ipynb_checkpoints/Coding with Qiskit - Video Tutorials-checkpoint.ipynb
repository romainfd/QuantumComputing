## Hello World
[Hello World — Programming on Quantum Computers Ep 3](https://www.youtube.com/watch?v=RrUTwq5jKM4&list=PLOFEBzvs-Vvp2xg9-POLJhQwtVktlYGbY&index=3)

Creating a quantum circuit and simulating it

```python
from qiskit import *
```

```python
qr = QuantumRegister(2)
```

```python
cr = ClassicalRegister(2)
```

```python
circuit = QuantumCircuit(qr, cr)
```

```python
%matplotlib inline
```

```python
circuit.draw()
```

```python
circuit.h(qr[0])
circuit.draw(output='mpl')
```

```python
circuit.cx(qr[0], qr[1])
circuit.draw(output='mpl')
```

```python
circuit.measure(qr, cr)
circuit.draw(output='mpl')
```

### Simulation

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(circuit, backend=simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(result.get_counts(circuit))
```

### Real Quantum Computer run

```python
IBMQ.load_account()
```

```python
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_london')
```

```python
job = execute(circuit, backend=qcomp)
```

```python
from qiskit.tools.monitor import job_monitor
```

```python
job_monitor(job)
```

```python
result = job.result()
plot_histogram(result.get_counts(circuit))
```

## Building blocks
[Building Blocks of Quantum Circuits — Programming on Quantum Computers Ep 4](https://www.youtube.com/watch?v=tBnWG_95F9c&list=PLOFEBzvs-Vvp2xg9-POLJhQwtVktlYGbY&index=4)

Illustration of the different representations of a circuit block through the example of the X-block

```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
```

### State Vector representation on Bloch sphere

```python
circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
statevector = result.get_statevector()
print(statevector)
%matplotlib inline
circuit.draw(output='mpl')
```

```python
plot_bloch_multivector(statevector)
```

### Measures representation

```python
circuit.measure([0], [0])
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
plot_histogram(result.get_counts())
```

### Matrix representation

```python
circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend=simulator).result()
unitaryvector = result.get_unitary()
print(unitaryvector)
```

## Quantum Teleportation
[Quantum Teleportation Algorithm — Programming on Quantum Computers Ep 5
](https://www.youtube.com/watch?v=mMwovHK2NrE&list=PLOFEBzvs-Vvp2xg9-POLJhQwtVktlYGbY&index=5)

```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3, 3)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1, 2)
circuit.draw(output='mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0, 1], [0, 1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
circuit.measure(2, 2)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(counts)
```

## Bernstein-Vazirani algorithm
[Bernstein-Vazirani Algorithm — Programming on Quantum Computers Ep 6](https://www.youtube.com/watch?v=sqJIpHYl7oo&list=PLOFEBzvs-Vvp2xg9-POLJhQwtVktlYGbY&index=6)

Guessing a number in 1 try

```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secret = '10101'
secret_len = len(secret)
```

```python
circuit = QuantumCircuit(secret_len + 1, secret_len)
circuit.h(range(secret_len))
circuit.x(secret_len)
circuit.h(secret_len)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
for i, c in enumerate(reversed(secret)):
    if c == '1':
        circuit.cx(i, secret_len)
circuit.barrier()
circuit.h(range(secret_len))
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure(range(secret_len), range(secret_len))
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1).result()
print(result.get_counts())
```
